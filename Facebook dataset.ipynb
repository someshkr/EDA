```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sex = pd.get_dummies(df['gender'],drop_first=True)
```

```python
sex.head()
```

```python
df.info()
```

```python
df.dropna() 
```

```python
df = pd.concat([df,sex], axis = 1)
```

```python
df.head()
```

```python
#df.drop(['gender'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
labels = ['Teen','20s','30s','40s','50s','Over 50s']
df['age_group'] = pd.cut(df.age,bins = [12,19,29,39,49,59,df.age.max()],
                         labels= labels,include_lowest = True)
df.head(10)
```

```python
age_grp = df.groupby('age_group')['tenure'].count()
fig,ax = plt.subplots(figsize = (15,8))
explode = [0.05,0.05,0.05,0.054,0.057,0.059]
age_grp.plot.pie(ax= ax,autopct = '%0.2f%%',explode=explode)
```

```python
gender_grp = df.groupby(['age_group','gender'])['tenure'].count()
fig,ax = plt.subplots(figsize = (15,7))
explode= [0.01,0.02,0.025,0.028,0.06,0.062,0.035,0.045,0.043,0.009,0.085,0.007]
gender_grp.plot.pie(ax= ax,autopct = '%0.2f%%',explode = explode)
plt.title('Age group vs gender distribution of  facebook users ')
```

```python
#gender and age group vs likes received distribution of facebook users

fig,ax = plt.subplots(figsize = (15,9))
color = ['g','b']
test = df.pivot_table('likes_received',index = 'age_group',columns = 'gender'
                     ,aggfunc='count')

for col in test.columns:
    test[col] = test[col]/sum(test[col])*100
test.plot(kind = 'bar', color = color, ax = ax,alpha= 0.7)

ax.set_xticklabels(test.index,rotation = 360)
ax.set_xlabel("Age group",fontsize=14)
ax.set_ylabel("Percentage of likes received",fontsize=14)
ax.set_title('Age and Gender Distribution of Facebook Users',fontsize=14)
```
